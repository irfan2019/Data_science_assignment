The function definition simply multiples the employee count with the share for each areas in order to get the number of employees for that particular region
- The used columns
    - EMP_ID
    - Location (Area name)
    - Lat (I have added a small noise tothe latitude, so it doesnt look like 10 employees in Barmbek live at the same house)
    - Long (I added a noise here to, to show variataions in the living location)

```python
def syn_data_gen(areas,employee_count):
    employee_dets = []
    for area, details in living_location.items():
        cnt = int(employee_count * details["share"])
        for idx in range(cnt):
            employee_dets.append({"EMP_ID": f"JNJMED_{len(employee_dets)+1}",
                                  "Location" : area,
                                  "Lat": details["lat"] + np.random.normal(0,0.02),
                                  "Long" : details["long"]  + np.random.normal(0,0.03)
                                  })
    syn_df = pd.DataFrame(employee_dets)
    return syn_df
```
